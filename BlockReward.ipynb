````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
````

````python
df = pd.read_csv("block_data.csv")
print(df)
````

````python
%matplotlib inline
plt.xlabel('timestamp(seconds)')
plt.ylabel('blockreward(ether)')
plt.scatter(df.timeStamp,df.blockReward,marker=".")
````

````python
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(df.timeStamp.values.reshape(-1,1),df.blockReward.values.ravel())
````

````python
new_timeStamp=np.array([[1686534888]])
predicted_blockReward=model.predict(new_timeStamp)

plt.scatter(df.timeStamp,df.blockReward,label='Data Points')
plt.scatter(new_timeStamp,predicted_blockReward,color="red",label='Predicted Future Block Reward')
plt.xlabel('TimeStamp')
plt.ylabel('Block Reward')
plt.legend()
plt.show()

print("Predicated Block Reward:",predicted_blockReward[0])
````
